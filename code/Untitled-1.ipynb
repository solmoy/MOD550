```python


import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Load noisy oscillatory data
x, y = np.load('oscillatory_data.npy', allow_pickle=True)

#  Convert to PyTorch tensors
x_tensor = torch.tensor(x, dtype=torch.float32).view(-1, 1)
y_tensor = torch.tensor(y, dtype=torch.float32).view(-1, 1)

# Normalize the data
scaler_x = StandardScaler()
scaler_y = StandardScaler()
x_scaled = torch.tensor(scaler_x.fit_transform(x_tensor), dtype=torch.float32)
y_scaled = torch.tensor(scaler_y.fit_transform(y_tensor), dtype=torch.float32)


# Define Fully Connected Neural Network (FCN)
class FCN(nn.Module):
    def __init__(self, N_INPUT, N_OUTPUT, N_HIDDEN, N_LAYERS, activation=nn.Tanh):
        super().__init__()
        self.fcs = nn.Sequential(nn.Linear(N_INPUT, N_HIDDEN), activation())
        self.fch = nn.Sequential(*[
            nn.Sequential(nn.Linear(N_HIDDEN, N_HIDDEN), activation()) 
            for _ in range(N_LAYERS - 1)
        ])
        self.fce = nn.Linear(N_HIDDEN, N_OUTPUT)

    def forward(self, x):
        x = self.fcs(x)  
        x = self.fch(x)  
        x = self.fce(x)  
        return x


# Define FCN model for Regression
model = FCN(N_INPUT=1, N_OUTPUT=1, N_HIDDEN=64, N_LAYERS=3, activation=nn.Tanh)

# Define optimizer and loss function
optimizer = optim.Adam(model.parameters(), lr=0.005)
criterion = nn.MSELoss()

# Train the model
epochs = 5000
losses = []

for epoch in range(epochs):
    optimizer.zero_grad()
    y_pred = model(x_scaled)  
    loss = criterion(y_pred, y_scaled)  
    loss.backward()
    optimizer.step()
    losses.append(loss.item())

    if epoch % 500 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.6f}")

# Plot training loss
plt.figure(figsize=(6, 3))
plt.plot(losses, label="Loss over Iterations")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("PINN Training Loss")
plt.legend()
plt.show()

# Predict using the trained model
y_pred_scaled = model(x_scaled).detach()
y_pred = scaler_y.inverse_transform(y_pred_scaled.numpy())  # Convert back to original scale

# Plot FCN Regression Results
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Actual Data', alpha=0.3)
plt.scatter(x, y_pred, color='red', label='PINN Prediction', linestyle="dashed")
plt.title('Physics-informed neural network Regression')
plt.xlabel('Feature (x)')
plt.ylabel('Output (y)')
plt.legend()
plt.show()

print("Task completed: PINN")
```
